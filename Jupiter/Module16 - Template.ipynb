<h1>What is the average time it takes for a bus to complete the route</h1>

```python
#import librairies
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import numpy as np
from haversine import haversine
import pandas as pd
from pyfiglet import Figlet #for fun
figlet=Figlet(font='small')
```

```python
#list of busses
bus_list=df['label'].unique()
bus_list
```

```python
#Loop function to find when buch rearrives at stop
def route_times(): 
    route_time_dict = {}
    for i in range(0,len(bus_list)):    
        bus = df[df['label'] == bus_list[i]] #select bus
        stops = bus['current_stop_sequence']
        first_stop = stops.iloc[0]
        first_stop_time = bus['updated_at'].iloc[0]
        next = False 
        for j in range(0, len(stops)):
            if stops.iloc[j] != first_stop: 
                next = True
            if stops.iloc[j] == first_stop and next == True:
                last_stop_time = bus['updated_at'].iloc[j]
                route_time_dict[bus_list[i]] = \
                    datetime.strptime(last_stop_time, '%Y-%m-%d %H:%M:%S') -\
                    datetime.strptime(first_stop_time, '%Y-%m-%d %H:%M:%S')
                break 
    return route_time_dict

route_times()
print(figlet.renderText('Route Time for Each Bus :'))
print(route_times()) 
```

```python
route_times_values = pd.Series(list(route_times().values()), index = route_times().keys())
mean_time_mins= round((route_times_values.mean()) / timedelta(minutes=1),2)
print(figlet.renderText('Average Route Time:'))
print(f'{mean_time_mins} minutes')
```

<h1>Provide a Visualization based on the data</h1>

```python
#Plotting the route time for each bus
route_times_mins = route_times_values / timedelta(minutes=1)
busList = list(route_times().keys())
plt.bar(busList, route_times_mins, color='g', width=.8, align='edge')
plt.title('Minutes to Complete Route')
plt.xlabel('Bus')
plt.ylabel('Minutes')
plt.show()
```

```python
#slightly clearer as scatter plot
plt.scatter(busList, route_times_mins)
```

<h1>Give an estimate of the speed of the bus from current_stop_sequence = 1 to the last  current_stop_sequence. </h1> 

Note: You can use the haversine Python library to calculate the distance between two points, given two longitude and latitude coordinates.

```python
#haversine used for distance between of first and last location on route 1
loc_1=(42.32983,-71.08371)
loc_24=(42.37270345,-71.11644567)
kilometers=haversine(loc_1,loc_24)
#speend calculated by the above distance / average time from above
KpH=round(kilometers/(mean_time_mins/60),2)
print(figlet.renderText(f'Average Speed is :'))
print(f'{KpH} km/h') 

```
